```python
import json
from pathlib import Path
from agentic_data.evals import eval_batch
import pandas as pd
import nest_asyncio
nest_asyncio.apply()

FOLDER_NAME = Path.cwd() / "runs"

def load_run(run_name: str):
    run_file = FOLDER_NAME / f"{run_name}.json"
    with open(run_file, 'r') as f:
        data = json.load(f)
    return data
```

**全流程混合数据集评测**

```python
common_run = load_run("common_20260227_124248")
report_common_dataset = pd.DataFrame(await eval_batch(common_run))
report_common_dataset
```

```python
report_common_dataset = report_common_dataset[report_common_dataset.reasoning.notnull()].copy()
print("Evaluation scores:")
scores = ['overall_score', 'workflow_score', 'robotic_score',
       'naturalness_score',  'deviation_score', 'handoff_score'
]
report_common_dataset[scores].mean()
```

```python
handoff_run = load_run("handoff_20260227_124319")
report_handoff_dataset = pd.DataFrame(await eval_batch(handoff_run))
report_handoff_dataset
```

```python
report_handoff_dataset = report_handoff_dataset[report_handoff_dataset.reasoning.notnull()].copy()
print("Evaluation scores:")
scores = ['overall_score', 'workflow_score', 'robotic_score',
       'naturalness_score',  'deviation_score', 'handoff_score'
]
report_handoff_dataset[scores].mean()
```
